#### MENOR DE DOS PARES: Escriba una función que devuelva el menor de dos números dados *si* ambos números son pares, pero devuelva el mayor si uno o ambos números son impares
    lesser_of_two_evens(2,4) --> 2
    lesser_of_two_evens(2,5) --> 5

```python
# Esta función recibe dos números y retorna el menor si ambos son pares, o el mayor si alguno es impar
def lesser_of_two_evens(a,b):

    # Verifica si ambos son pares tin
    if a % 2 == 0 and b % 2 == 0:
        # Retorna el menor de los dos
        return min(a, b)
    
    else:
        # Retorna el mayor si alguno es impar
        return max(a, b)
```

```python
# Revisión
lesser_of_two_evens(2,4)
```

```python
# Revisión
lesser_of_two_evens(2,5)
```

#### GALLETAS DE ANIMALES: Escribir una función que toma una cadena de dos palabras y devuelve True si ambas palabras comienzan con la misma letra
    animal_crackers('Ciudad Calor') --> True
    animal_crackers('Lugar Frio') --> False

```python
# Esta función recibe una cadena de dos palabras y retorna True si ambas comienzan con la misma letra
def animal_crackers(text):
    # Divide el texto en palabras
    palabras = text.split()
    # Compara las primeras letras en minúsculas
    return palabras[0][0].lower() == palabras[1][0].lower()
```

```python
# Revisión
animal_crackers('Ciudad Calor')
```

```python
# Revisión
animal_crackers('Lugar Frio')
```

#### HACE VEINTE: dados dos enteros, devuelve True si la suma de los enteros es 20 o si uno de los enteros es 20. De lo contrario, devuelve False

    makes_twenty(20,10) --> True
    makes_twenty(12,8) --> True
    makes_twenty(2,3) --> False

```python
# Esta función recibe dos enteros y retorna True si su suma es 20 o si uno de ellos es 20
def makes_twenty(n1,n2):
    # Verifica si la suma es 20 o si alguno es 20
    return n1 + n2 == 20 or n1 == 20 or n2 == 20
```

```python
# Revisión
makes_twenty(20,10)
```

```python
# Revisión
makes_twenty(2,3)
```

```python
makes_twenty(12,8)
```

#### MAYÚSCULAS: Escribir una función que haga mayúscula la primera y cuarta letra de una cadena de texto
     
    old_macdonald('macdonald') --> MacDonald
    
Note: `'macdonald'.capitalize()` returns `'Macdonald'`

```python
# Esta función recibe una cadena y retorna la cadena con la primera y cuarta letra en mayúscula
def old_macdonald(name):
    # Verifica si tiene al menos 4 caracteres
    if len(name) > 3:
        # Capitaliza la primera y cuarta letra
        return name[0].upper() + name[1:3] + name[3].upper() + name[4:]
    
    else:
        # Solo capitaliza la primera si es muy corta
        return name.capitalize()
```

```python
# Check
old_macdonald('macdonald')
```

#### REVERSA: Dada una oración, retornar la misma oración pero en reversa

    master_yoda('I am home') --> 'home am I'
    master_yoda('We are ready') --> 'ready are We'
    
Note: Revisar método .join()

    >>> "--".join(['a','b','c'])
    >>> 'a--b--c'

Resultado de .join():

    >>> " ".join(['Hello','world'])
    >>> "Hello world"

```python

# Esta función recibe una oración y retorna las palabras en orden inverso
def master_yoda(text):
    # Divide la oración en palabras
    palabras = text.split()
    
    # Une las palabras en orden inverso
    return ' '.join(reversed(palabras))
```

```python
# Revisión
master_yoda('I am home')
```

```python
# Revisión
master_yoda('We are ready')
```

#### PROBLEMA 33: 

Dada una lista de enteros, retornar True si el arreglo tinene un  3  seguido por un  3 .

    has_33([1, 3, 3]) → True
    has_33([1, 3, 1, 3]) → False
    has_33([3, 1, 3]) → False

```python
# Esta función recibe una lista de enteros y retorna True si hay un 3 seguido de otro 3
def has_33(nums):
    
    # Recorre la lista hasta el penúltimo elemento
    for i in range(len(nums) - 1):
        # Verifica si hay dos 3 consecutivos
        if nums[i] == 3 and nums[i + 1] == 3:
            # Retorna True si los encuentra
            return True
        
    # Retorna False si no los encuentra
    return False
```

```python
# Revisión
has_33([1, 3, 3])
```

```python
# Revisión
has_33([1, 3, 1, 3])
```

```python
# Revisión
has_33([3, 1, 3])
```

#### REPLICADOR: Dada una cadena, devuelve una cadena donde por cada carácter en el original hay tres caracteres REPETIDOS
    paper_doll('Hello') --> 'HHHeeellllllooo'
    paper_doll('Mississippi') --> 'MMMiiissssssiiippppppiii'

```python
# Esta función recibe una cadena y retorna cada carácter repetido 3 veces
def paper_doll(text):
    
    # Inicializa cadena vacía
    resultado = ''
    
    # Recorre cada carácter
    for char in text:
        # Agrega el carácter 3 veces
        resultado += char * 3
    
    # Retorna el resultado
    return resultado
```

```python
# Check
paper_doll('Hello')
```

```python
# Check
paper_doll('Mississippi')
```

#### BLACKJACK: Dados tres enteros entre 1 y 11, si su suma es menor o igual a 21, devuelve su suma. Si su suma excede 21 *y* hay un once, reduce la suma total en 10. Finalmente, si la suma (incluso después del ajuste) excede 21, devuelve 'BUST'
    blackjack(5,6,7) --> 18
    blackjack(9,9,9) --> 'BUST'
    blackjack(9,9,11) --> 19

```python
# Esta función recibe tres enteros (1-11) y retorna su suma si es ≤21, ajusta si hay 11 y suma >21, o 'BUST' si excede 21
def blackjack(a,b,c):

    # Calcula la suma total d elos números q nos dan
    suma = a + b + c

    # Si la suma no excede 21
    if suma <= 21:
        # Retorna la suma
        return suma
    
    # Si hay un 11 y al restar 10 no excede 21
    elif 11 in [a, b, c] and suma - 10 <= 21:
        # Retorna la suma ajustada
        return suma - 10
    
    # Retorna 'BUST' si excede 21
    else:
        return 'BUST'
```

```python
# Revisión
blackjack(5,6,7)
```

```python
# Revisión
blackjack(9,9,9)
```

```python
# Revisión
blackjack(9,9,11)
```
